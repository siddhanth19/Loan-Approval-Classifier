```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train=pd.read_csv("Desktop//csv files//Loan Approval Classifier//train.csv")
test=pd.read_csv("Desktop//csv files//Loan Approval Classifier//test.csv")
```

```python
train.isnull().sum()
```

```python
train.describe()
```

```python
train.head()
```

```python
train.drop(['Loan_ID','Gender'],axis='columns',inplace=True)
```

```python
import numpy as np
```

```python
train.shape
```

```python
train.Dependents=train.Dependents.fillna(method='ffill')
```

```python
train.LoanAmount.fillna(train.LoanAmount.mean(),inplace=True)
train.Loan_Amount_Term.fillna(train.Loan_Amount_Term.mean(),inplace=True)
```

```python
train.dropna(inplace=True)
```

```python
test.Dependents.fillna(method='ffill')
test.LoanAmount.fillna(train.LoanAmount.mean(),inplace=True)
test.Loan_Amount_Term.fillna(train.Loan_Amount_Term.mean(),inplace=True)
test.dropna(inplace=True)
id=test['Loan_ID']
test.drop(['Loan_ID','Gender'],axis='columns',inplace=True)
```

```python
x=train.drop(['Loan_Status'],axis='columns')
y=train['Loan_Status']
```

```python
train_x=pd.get_dummies(x)
test_x=pd.get_dummies(test)
train_x
```

```python
train_x.drop(['Married_No','Dependents_3+','Education_Not Graduate','Self_Employed_No','Property_Area_Rural'],axis='columns',inplace=True)
test_x.drop(['Married_No','Dependents_3+','Education_Not Graduate','Self_Employed_No','Property_Area_Rural'],axis='columns',inplace=True)
```

```python
train_x
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sk=StandardScaler()
x_scaled=sk.fit_transform(train_x)
x_scaled
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=1000)
```

```python
model.fit(x_scaled,y)
```

```python
test_x_scaled=sk.fit_transform(test_x)
```

```python
predicted=model.predict(test_x_scaled)
```

```python
model.score(x_scaled,y)
```

```python
predicted
```

```python
result=pd.DataFrame(predicted,index=id,columns=['Loan_Status'])
result
```

```python
result.to_csv("Desktop//csv files//Loan Approval Classifier//Submission.csv")
```
